### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    # YOUR CODE HERE
    try:
        return a / b
    except ZeroDivisionError:
        return None
    except TypeError:
        raise TypeError("Inputs must be numeric")
    pass

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A) # YOUR CODE HERE

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    # Left hand side and Right hand side
    lhs = A @ v
    rhs = lam * v

    is_equal = np.allclose(lhs, rhs)
    print(f"\nEigenpair {i+1} (λ = {lam}):")
    print(f"  A @ v: {lhs}")
    print(f"  λ * v: {rhs}")
    print(f"  Verified? {is_equal}")

# YOUR CODE HERE

# Plot
plt.figure(figsize=(6, 6))

# Origin point
origin = [0, 0]

colors = ['crimson', 'dodgerblue']
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]
    scaled_v = lam * v

    # Plot original unit eigenvector
    plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1,
               color=colors[i], alpha=0.4, label=f'v{i+1} (Unit)')
    # Plot scaled eigenvector
    plt.quiver(*origin, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1,
               color=colors[i], label=f'λ{i+1} * v{i+1} (Scaled by {lam})')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(True, linestyle=':', alpha=0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Eigenvectors Scaled by Eigenvalues')
plt.legend()
plt.gca().set_aspect('equal')

plt.show()

# YOUR CODE HERE

print("4.3 passed ✓")
````
