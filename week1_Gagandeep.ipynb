### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
import numpy as np
from scipy import stats
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# 1. KS test — compare each against N(0,1)
# We pass 'norm' to compare against a standard normal distribution CDF
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# 2. ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# 3. Difference once and re-run
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    epsilon = 1e-10

    # 1. Determine common bin edges based on the expected (baseline) data
    percentiles = np.linspace(0, 100, bins + 1)
    bin_edges = np.percentile(expected, percentiles)

    # Adjust outer edges slightly to ensure all data points fall inside
    bin_edges[0] -= 1e-5
    bin_edges[-1] += 1e-5

    # 2. Count frequencies in each bin
    expected_counts, _ = np.histogram(expected, bins=bin_edges)
    actual_counts, _ = np.histogram(actual, bins=bin_edges)

    # 3. Convert counts to proportions (percentages)
    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    # 4. Handle zeros using epsilon to avoid division by zero or log(0)
    expected_pct = np.where(expected_pct == 0, epsilon, expected_pct)
    actual_pct = np.where(actual_pct == 0, epsilon, actual_pct)

    # 5. Compute individual bin PSI and sum them up
    psi_values = (actual_pct - expected_pct) * np.log(actual_pct / expected_pct)
    return np.sum(psi_values)

# Setup simulated data
np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

# Compute value
psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots(figsize=(7, 4))
sns.histplot(train_dist, color="blue", label="Training Distribution", kde=True, ax=ax, stat="probability", alpha=0.4, bins=25)
sns.histplot(drift_dist, color="red", label="Production Distribution", kde=True, ax=ax, stat="probability", alpha=0.4, bins=25)

ax.set_xlabel('Feature Value')
ax.set_ylabel('Probability')
ax.legend()
ax.set_title('Distribution Shift Visualization')
plt.tight_layout()
plt.show()
````

Concept Drift: A change over time in the statistical relationship between the model input features and the target variable (shifts), meaning the underlying logic of what makes a prediction "correct" has changed.


Covariate Drift (Data Drift): A shift in the independent distribution of the model's input features over time (shifts), while the true relationship between those features and the final target remains unchanged.

- 0.1<= psi <= 0.2 :- Minor shift; implies moderate change and requires close monitoring or proactive review.

## Part 6 — Probability Theory

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

# Sample Space: S = {4 Red, 3 Blue, 3 Green} with total elements = 10
p_red   = red / total
p_blue  = blue / total
p_green = green / total

# Conditional: P(blue second | red first)
# After drawing 1 red, 3 blue marbles remain out of 9 total marbles
p_blue_given_red = blue / (total - 1)

# Joint: P(red first AND blue second)
# Formula: P(A and B) = P(A) * P(B|A)
p_red_then_blue = p_red * p_blue_given_red

# Independence check: are P(B|A) == P(B) ?
independent = (p_blue_given_red == p_blue)

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# 1. P(free) — law of total probability
# Formula: P(word) = P(word|Spam)*P(Spam) + P(word|Ham)*P(Ham)
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)

# 2. Bayes Theorem Calculation
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

# 3. Functional implementation
def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    prior_ham = 1.0 - prior_spam
    evidence = (p_word_given_spam * prior_spam) + (p_word_given_ham * prior_ham)
    posterior_spam = (p_word_given_spam * prior_spam) / evidence
    return posterior_spam

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(7)
# 1. Define a population following an exponential distribution
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# 2. Repeatedly draw samples of size n=30 and record their means 5,000 times

sample_indices = np.random.choice(population, size=(n_samples, sample_size))
sample_means = np.mean(sample_indices, axis=1)

pop_mean = population.mean()
pop_std  = population.std()

# 3. Calculate predicted standard error by CLT: σ / sqrt(n)
clt_std  = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# 4. Plot the histogram of sample means and overlay the theoretical normal distribution
fig, ax = plt.subplots(figsize=(7, 4))

# Empirical sample means histogram (normalized to a density plot)
count, bins, ignored = ax.hist(sample_means, bins=40, density=True, alpha=0.6, color='skyblue', label='Sample Means')

# Theoretical Normal Curve calculation
x_axis = np.linspace(bins.min(), bins.max(), 300)
theoretical_normal = stats.norm.pdf(x_axis, loc=pop_mean, scale=clt_std)
ax.plot(x_axis, theoretical_normal, color='darkorange', linewidth=2.5, label='CLT Theoretical Normal')

ax.set_title('CLT: Distribution of Sample Means')
ax.set_xlabel('Sample Mean Value')
ax.set_ylabel('Probability Density')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# 5. Run a KS test to confirm the distribution of sample means is approximately normal
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> *In 2–3 sentences: Why does this result matter for statistics and machine learning? What assumption does the CLT justify in many classical statistical tests?*

CLT Importance:-

Universal Approximation: The Central Limit Theorem (CLT) allows you to confidently model the behavior of sample averages as a normal distribution, even if the underlying source data is heavily skewed, binary, or completely unknown.

Predictable Uncertainty: It provides a reliable mathematical blueprint to calculate confidence intervals, determine minimal sample sizes, and estimate error margins for machine learning model performance evaluation.

Foundation for Ensembles: Algorithms like Bagging and Random Forests rely on the CLT; as you average predictions from thousands of independent, highly volatile individual trees, the variance of the overall ensemble stabilizes and drops significantly.



Assumptions:-


The Normality Assumption: The CLT directly justifies the assumption of normality for sampling distributions in classical inferential tests.

Sampling vs. Population: It proves that while your raw population data does not need to be normally distributed, the mean of that data will behave normally when sampled repeatedly.

Classical Test Safeguard: This explicit rule is what makes parametric frameworks like t-tests, ANOVA, and linear regression residual tests mathematically valid and accurate when applied to messy, non-normal real-world data.
